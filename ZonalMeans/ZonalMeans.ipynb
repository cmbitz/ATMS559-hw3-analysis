```python
%matplotlib inline
%load_ext autoreload
%autoreload
import matplotlib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import matplotlib.colors as colors

import numpy as np
import xarray as xr
```

```python
# plotting and averaging functions
def zonal1D(field, ax, title='',label=''):
    field.mean(dim='time').plot(ax=ax,label=label)
    ax.set_xticks(np.arange(-90, 120, 30))
    ax.set_title(title)
    ax.set_xticklabels(['90S','60S','30S','0','30N','60N','90N'])
    ax.set_xlabel('Latitude')


def zonalsigma(field, ax, title='', vmin=-20, vmax=20):
    field.mean(dim='time').plot.pcolormesh(ax=ax, vmin=vmin, vmax=vmax,cmap="coolwarm")
    #field.mean(dim='time').mean(dim='lon').plot.contourf(ax=ax, vmin=-vmax, vmax=vmax,cmap="coolwarm")

    ax.invert_yaxis()
    ax.set_xticks(np.arange(-90, 120, 30))
    ax.set_title(title)
    ax.set_xticklabels(['90S','60S','30S','0','30N','60N','90N'])
    ax.set_xlabel('Latitude')
    ax.set_ylabel('Pressure')
    
def ZonalMeanCombineFieldNicely(ddir,casenames,prefernames,thevar):  # combine the fields for one variable into a nice xarray DataArray
    x_all=[]
    for i,casename in enumerate(casenames):
        ncfile=ddir+casename+'_'+thevar+'_climo.nc'
        x=xr.open_dataset(ncfile)[thevar]
        x['time'] = np.arange(1,13) # fix some of my times
        x=x.mean(dim='lon') 
        x = x.assign_coords(casename=prefernames[i])
        x = x.expand_dims('casename')
        x.name=thevar
        x_all.append(x)
    x_all=xr.concat(x_all, "casename")
    x_all
    return x_all

def compute_global(da,gw): # multiply by so-called Gaussian weights that oddly enough sum to 2
    globalmean=(da.mean(dim='lon')*gw).sum(dim='lat')/2
    return globalmean

# get Gaussian Weights, gw, from any old cam hist file, a bit slow, sorry
gw=xr.open_dataset('/glade/derecho/scratch/bitz/archive/AMIPpiControl/atm/hist/AMIPpiControl.cam.h0.0001-01.nc').gw
```

```python
# Fixed SST runs
ddirSST = '/glade/derecho/scratch/bitz/ADF/climo/FixedSST/'  # this is where the climatologies are

# these are the casenames used for FixedSST runs
casenamesSST = ["AMIPpiControl",
            "DoubleCO2FixedSST",
            "QuadrupleCO2FixedSST",
            "AlotofCH4FixedSST",
            "RaiseSolarFixedSST",
            "solar_decrease1",
            "DirtyAirFixedSST",
            "IcyNAFixedSST",
            "DarkSeaIceFixedSST",
            "NoTreesFixedSST",
            "fixedSSTFlatTibet",
            "FlatAntarcticaFixedSST",
            "FlatRockiesFixedSST"]

# Fully Coupled 
ddir = '/glade/derecho/scratch/bitz/ADF/climo/FullyCoupled/'  # this is where the climatologies are

# these are the casenames used for fully coupled runs #  "b.e21.B1850.f19_g17.CMIP6-piControl-2deg.001",
casenames = ["piControl", 
            "DoubleCO2",
            "b.e21.BCO2x4.f19_g17.CMIP6-abrupt4xCO2-2deg.001",
            "AlotofCH4",
            "RaiseSolar",
            "LowerSolar",
            "DirtyAir",
            "IcyNA",
            "DarkIce",
            "NoTrees",
            "FlatTibet",
            "FlatAntarctica",
            "FlatRockies",
            "b.e21.B1850.f19_g17.CMIP6-piControl-2deg.001"]

# this is what they were meant to be named
prefernames = ["piControl",
            "DoubleCO2",
            "QuadrupleCO2",
            "AlotofCH4",
            "RaiseSolar",
            "LowerSolar",
            "DirtyAir",
            "IcyNA",
            "DarkSeaIce",
            "NoTrees",
            "FlatTibet",
            "FlatTibet-Antarctica",
            "FlatTibet-Antarctica-Rockies",
            "CMIP6piControl"]
```

## I'll demonstrate some possibilities here. There are many more you can try on our own

```python
# Get a field from a single CAM variable like trefht
trefht_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'TREFHT')
print('\n\nYou just acquired all TREFHT from Fixed SST runs combined together!\n\n')
#print(trefht_all_SST) # unccomment to see the DataArray
```

```python
# Get a field by combininb CAM variable like TOM imbalance, we'll difference it from piControl to get ERF
fsnt_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'FSNT')
flnt_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'FLNT')
imbalance_all_SST=fsnt_all_SST-flnt_all_SST
imbalance_all_SST.name='imbalance'
#print(imbalance_all) # uncomment if you want to see the DataArray 
```

```python
# Get a field by combining CAM variable like total precipitation
precl_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'PRECL')
precc_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'PRECC')
prect_all_SST=precl_all_SST+precc_all_SST
prect_all_SST.name='prect'
```

```python
T_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'T')
T_all_SST.name='T(z)'
```

```python
U_all_SST=ZonalMeanCombineFieldNicely(ddirSST,casenamesSST,prefernames,'U')
```

```python
# now let's get a few of the fully coupled run variables 

# Get a field from a single CAM variable like trefht
trefht_all=ZonalMeanCombineFieldNicely(ddir,casenames,prefernames,'TREFHT')
T_all=ZonalMeanCombineFieldNicely(ddir,casenames,prefernames,'T')
U_all=ZonalMeanCombineFieldNicely(ddir,casenames,prefernames,'U')

# Get a field by combining CAM variable like total precipitation
precl_all=ZonalMeanCombineFieldNicely(ddir,casenames,prefernames,'PRECL')
precc_all=ZonalMeanCombineFieldNicely(ddir,casenames,prefernames,'PRECC')
prect_all=precl_all+precc_all
prect_all.name='prect'
```

```python
# leave off the Flat Mts since they get lost
fig, ax = plt.subplots( 1, 1, figsize=(8, 5) )
for i,casename in enumerate(prect_all_SST.casename[1:10]):
    field = trefht_all_SST.sel(casename=casename)-trefht_all_SST.sel(casename='piControl')
    zonal1D(field, ax, title='',label=casename.values) # convert m/s to cm/day
plt.ylabel('$\Delta$ trefht - K')   
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.title('Fixed SST Perturbed minus Control' )
```

```python
# leave off the Flat Mts since they get lost
fig, ax = plt.subplots( 1, 1, figsize=(8, 5) )
for i,casename in enumerate(prect_all_SST.casename[1:10]):
    field = prect_all_SST.sel(casename=casename)-prect_all_SST.sel(casename='piControl')
    zonal1D(field*8640000, ax, title='',label=casename.values) # convert m/s to cm/day
plt.ylabel('$\Delta$ Precipitation - cm per day')   
plt.title('Fixed SST Perturbed minus Control')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
```
